```python
from bs4 import BeautifulSoup
import requests
```

```python
df = pd.read_pickle("Data/netflix-complete-data-w-genre.pickle")
```

```python
imdbids = df["imdbid"].sample(n=5, random_state=2)
```

```python
for imdbid in imdbids:
    page = requests.get(f"https://www.imdb.com/title/{imdbid}/ratings")
    soup = BeautifulSoup(page.content, "html.parser")
    st = soup.find_all("div", attrs={"class":"allText"})[0].text.strip()
    st_lower = st[:st.find("IMDb")].strip().replace(",", "")
    print(st_lower)
```

## Add number of votes to the main df

```python
def get_votes(row):
    imdbid = row["imdbid"]
    page = requests.get(f"https://www.imdb.com/title/{imdbid}/ratings")
    soup = BeautifulSoup(page.content, "html.parser")
    try:
        st = soup.find_all("div", attrs={"class":"allText"})[0].text.strip()
        st_lower = st[:st.find("IMDb")].strip().replace(",", "")
        return int(st_lower)
    except:
        return 0
```

```python
from tqdm import tqdm
tqdm.pandas()

df["votes"] = df.progress_apply(lambda row: get_votes(row) if row["imdbid"] else 0, axis=1)
```

```python
df.to_pickle("netflix_pull_w_votes.pickle")
```
